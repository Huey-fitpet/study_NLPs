## hugging face 
- 여러 회사들이 개발한 모델이나 제공하는 데이터셋을 편리하게 사용하는 인터페이스 제공

```python
from transformers import AutoModel, AutoTokenizer
```

### BERT 모델 활용 준비
- https://huggingface.co/google-bert/bert-base-uncased

```python
AutoModel.from_pretrained('google-bert/bert-base-uncased')
```

```python
AutoTokenizer.from_pretrained('google-bert/bert-base-uncased')
```

```python
bert_model = AutoModel.from_pretrained('google-bert/bert-base-uncased')
bert_tokenizer = AutoTokenizer.from_pretrained('google-bert/bert-base-uncased')
```

```python
type(bert_model), type(bert_tokenizer)
```

```python
text = f'What is Hugging Face Transformer ?'
encoded_input = bert_tokenizer(text, return_tensors='pt') # 학습 완료된 형태의 벡터로 결과 리턴
bert_out = bert_model(**encoded_input)
```

```python
type(encoded_input), encoded_input
```

```python
type(bert_out), bert_out
```

## 모델 불러오기
- 모델 : 바디와 헤드로 구분

```python
# body only model 
# https://huggingface.co/klue/roberta-base
roberta_model = AutoModel.from_pretrained('klue/roberta-base')
```

```python
type(roberta_model), roberta_model
```

```python
from transformers import AutoModelForSequenceClassification
# body + head model 
# https://huggingface.co/SamLowe/roberta-base-go_emotions
emotion_model = AutoModelForSequenceClassification.from_pretrained('SamLowe/roberta-base-go_emotions')
```

```python
type(emotion_model), emotion_model
```

```python
# head 가 없는 모델을 목적에 의해 불러오기 
roberta_classification_model = AutoModelForSequenceClassification.from_pretrained('klue/roberta-base')
```

```python
type(roberta_classification_model), roberta_classification_model
```
