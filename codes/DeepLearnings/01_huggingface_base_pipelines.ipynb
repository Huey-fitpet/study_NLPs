```python
from transformers import pipeline

# 감정 분석을 위한 파이프라인 생성
classifier = pipeline("sentiment-analysis")

# 감정 분석 실행
results = classifier(["I love programming!", "This is a boring task."])
for result in results:
    print(f"Label: {result['label']}, Score: {result['score']:.4f}")
```

```python
generator = pipeline("text-generation", model="gpt2") # 텍스트 생성
text = generator("Once upon a time", max_length=50, num_return_sequences=2)
for idx, seq in enumerate(text):
    print(f"Generated #{idx + 1}: {seq['generated_text']}")
```

```python
text = generator("Warning!!!: ", max_length=50, num_return_sequences=2)
for idx, seq in enumerate(text):
    print(f"Generated #{idx + 1}: {seq['generated_text']}")
```

```python
qa_pipeline = pipeline("question-answering")

context = """ The Transformers library provides thousands of
pretrained models to perform tasks on texts such as classification,
information extraction, question answering, and more. """

question = "What does the Transformers library provide?"

result = qa_pipeline(question=question, context=context)
print(f"Answer: {result['answer']}")
```

```python
translator = pipeline("translation_en_to_fr")

text = "Transformers library is very powerful."

translated = translator(text)
print(f"Translated: {translated[0]['translation_text']}")
```
