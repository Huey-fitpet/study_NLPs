## 빈도수에 의한 자연어 처리

### 데이터 전처리

```python
it_sentences = [
    "Artificial intelligence technology is increasingly being applied to various sectors of society",
    "Cloud computing is making data storage and processing more efficient and is revolutionizing the IT infrastructure of organizations.",
    "Machine learning and deep learning algorithms are revolutionizing data analysis and predictive modeling.",
    "Blockchain technology is helping to improve the transparency and security of financial transactions.",
    "Cybersecurity is emerging as a critical issue in the digital age and requires constant attention and improvement.",
    # "Agile methodologies in software development enable organizations to respond effectively to rapid market changes.",
    # "Data science and analytics play an important role in the business decision-making process and make organizations more competitive.",
    # "Internet of Things (IoT) technologies are transforming everyday life into a more convenient and connected place.",
    # "Virtual Reality (VR) and Augmented Reality (AR) technologies are providing new experiences in entertainment, education, healthcare, and more.",
    # "Python is a widely used language for a variety of programming tasks, including data science, web development, and automated scripting."
]
```

```python
# 하나의 묶음 문장 작성
text = ''.join(it_sentences)
text
```

```python
text_splits = text.split(sep=' ')
text_splits[:4], type(text_splits), len(text_splits)
```

### 단어(명사, 품사, 대명사 등등) 사전

```python
# 단어마다 빈도수
from collections import Counter
counter_text_splits = Counter(text_splits)
counter_text_splits.most_common()[:10] # get list
# 그냥 자르면 불용어 가 너무 많이 나옴 (and, is, of, the )
# 학습시에 그렇다고 뺄수 없음. 문장에서 순서가 중요하기 때문
```

### word cloud

```python
from wordcloud import WordCloud
```

```python
wordcloud_instance = WordCloud(background_color='white')
wordcloud_instance.generate_from_text(text=text)
```

```python
wordcloud_instance.words_ # 영어만 잘 됨
# 영어에서는 불용어 처리 알아서 해줌
# 빈도에 따른 확률 알려줌
```

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud_instance)
plt.axis('off')
plt.show()
```
